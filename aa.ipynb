```python
import pandas as pd
```

```python
data = pd.read_csv("wiki_1000.csv")
```

```python
import numpy as np
chunks = np.array_split(data, 5)
for i, chunk in enumerate(chunks):
    chunk.to_csv(f'allDf/chunk_{i+1}.csv')
```

```python
chunks = np.array_split(data, 5)
for i, chunk in enumerate(chunks):
    chunk = chunk.head(100)  # Take the first 100 rows of each chunk
    chunk.to_csv(f'allDf/chunk_{i+1}.csv')
```
